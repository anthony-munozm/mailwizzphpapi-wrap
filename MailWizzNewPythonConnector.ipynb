```python
def get_subscribers_all(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, payload):
    
    print(payload)
    
    url = "{}/subscribers/all".format(INTERNAL_SERVER)
    
    headers = {
        'public-key': PUBLIC_KEY,
        'private-key': PRIVATE_KEY,
        'Content-Type': "application/x-www-form-urlencoded",
        'api-url': API_URL,
        'cache-control': "no-cache",
        'Postman-Token': "b27bb318-7376-4d05-b0de-191cbcfa9dcd"
        }

    response = requests.request("POST", url, data=payload.strip(), headers=headers)

    json_data = json.loads(response.text)
    
    return json_data
```

```python
def get_subscribers_user_delete(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, payload):
    
    url = "{}/subscribers/user/delete".format(INTERNAL_SERVER)
    
    headers = {
        'public-key': PUBLIC_KEY,
        'private-key': PRIVATE_KEY,
        'Content-Type': "application/x-www-form-urlencoded",
        'api-url': API_URL,
        'cache-control': "no-cache",
        'Postman-Token': "b27bb318-7376-4d05-b0de-191cbcfa9dcd"
        }
    
    response = requests.request("POST", url, data=payload, headers=headers)

    json_data = json.loads(response.text)
    
    return json_data
```

```python
def iterate_and_cleaning_up(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, payload, total=0):
    
    subs = get_subscribers_all(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, payload)
    
    print(subs)
    
    print(payload)
            
    if subs["data"]["count"] is not None:
        
        print(subs)
    
        for item in subs["data"]["records"]:

            subscriber_uid = item["subscriber_uid"]

            new_payload = payload+"&subscriber_uid="+subscriber_uid

            get_subscribers_user_delete(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, new_payload)

        if subs["data"]["count"] is None:
                        
            return {"status":"success", "removed_element":total}

        if subs["data"]["count"] is not None:

            total = int(total) + int(subs["data"]["count"])

            print("Iterating...")

            iterate_and_cleaning_up(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, payload, total)

        else: 

            return {"status":"success", "removed_element":total}
    
    return {"status":"success", "removed_element":total}
```

```python
from flask import Flask, request

app = Flask(__name__)

@app.route("/lists/show", methods=['GET', 'POST'])
def get_lists_show_api():
    
    if request.method == 'GET':

        PUBLIC_KEY = request.args.get("public-key", None)
        PRIVATE_KEY = request.args.get("private-key", None)
        API_URL = request.args.get("api-url", None)
        payload = str(request.__dict__["environ"]["QUERY_STRING"])

        result = get_lists_show(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, payload)

        return jsonify(result)    
    
    else:
        
        PUBLIC_KEY = request.headers["public-key"]
        PRIVATE_KEY = request.headers["private-key"]
        API_URL = request.headers["api-url"]
        payload = str(request.form)
        
        values = list(request.form.values())
        keys = list(request.form.keys())
        
        params_string = ""
        for x, y in zip(values, keys):
            new_pair = "=".join([y,x+"&"])
            params_string = params_string + new_pair
            
        params_string = params_string[:-1]

        result = get_lists_show(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, payload=params_string)

        return jsonify(result)    
        
        



@app.route("/lists/create", methods=['GET', 'POST'])
def get_lists_create_api():
    
    if request.method == 'GET':

        PUBLIC_KEY = request.args.get("public-key", None)
        PRIVATE_KEY = request.args.get("private-key", None)
        API_URL = request.args.get("api-url", None)
        payload = str(request.__dict__["environ"]["QUERY_STRING"])

        result = get_lists_create(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, payload)

        return jsonify(result)    
    
    else:
                
        PUBLIC_KEY = request.headers["public-key"]
        PRIVATE_KEY = request.headers["private-key"]
        API_URL = request.headers["api-url"]
        payload = str(request.form)
        
        values = list(request.form.values())
        keys = list(request.form.keys())
        
        params_string = ""
        for x, y in zip(values, keys):
            new_pair = "=".join([y,x+"&"])
            params_string = params_string + new_pair
        
        params_string = params_string[:-1]

        result = get_lists_create(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, payload=params_string)

        return jsonify(result)    



@app.route("/subscribers/user/add", methods=['GET', 'POST'])
def get_subscribers_user_add_api():
    
    if request.method == 'GET':

        PUBLIC_KEY = request.args.get("public-key", None)
        PRIVATE_KEY = request.args.get("private-key", None)
        API_URL = request.args.get("api-url", None)
        payload = str(request.__dict__["environ"]["QUERY_STRING"])

        result = get_subscribers_user_add(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, payload)

        return jsonify(result)    
    
    else:
        
        PUBLIC_KEY = request.headers["public-key"]
        PRIVATE_KEY = request.headers["private-key"]
        API_URL = request.headers["api-url"]
        payload = str(request.form)
        
        values = list(request.form.values())
        keys = list(request.form.keys())
        
        params_string = ""
        for x, y in zip(values, keys):
            new_pair = "=".join([y,x+"&"])
            params_string = params_string + new_pair
        
        params_string = params_string[:-1]

        result = get_subscribers_user_add(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, payload=params_string)

        return jsonify(result)    
    
    
@app.route("/subscribers/user/delete", methods=['GET', 'POST'])
def get_subscribers_user_delete_api():

    if request.method == 'GET':

        PUBLIC_KEY = request.args.get("public-key", None)
        PRIVATE_KEY = request.args.get("private-key", None)
        API_URL = request.args.get("api-url", None)
        payload = str(request.__dict__["environ"]["QUERY_STRING"])

        result = get_subscribers_user_delete(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, payload)

        return jsonify(result)    
    
    else:
        PUBLIC_KEY = request.headers["public-key"]
        PRIVATE_KEY = request.headers["private-key"]
        API_URL = request.headers["api-url"]
        payload = str(request.form)
        
        values = list(request.form.values())
        keys = list(request.form.keys())
        
        params_string = ""
        for x, y in zip(values, keys):
            new_pair = "=".join([y,x+"&"])
            params_string = params_string + new_pair
        
        params_string = params_string[:-1]
        print(params_string)

        result = get_subscribers_user_delete(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, payload=params_string)

        return jsonify(result)    
    
@app.route("/subscribers/all", methods=['GET', 'POST'])
def get_subscribers_user_all_api():
    
    if request.method == 'GET':

        PUBLIC_KEY = request.args.get("public-key", None)
        PRIVATE_KEY = request.args.get("private-key", None)
        API_URL = request.args.get("api-url", None)
        payload = str(request.__dict__["environ"]["QUERY_STRING"])

        result = get_subscribers_all(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, payload)

        return jsonify(result)    
    
    else:
        
        PUBLIC_KEY = request.headers["public-key"]
        PRIVATE_KEY = request.headers["private-key"]
        API_URL = request.headers["api-url"]
        payload = str(request.form)
        
        values = list(request.form.values())
        keys = list(request.form.keys())
        
        params_string = ""
        for x, y in zip(values, keys):
            new_pair = "=".join([y,x+"&"])
            params_string = params_string + new_pair
        params_string = params_string[:-1]
        print(params_string)
    
        result = get_subscribers_all(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, payload=params_string)

        return jsonify(result)
    
@app.route("/lists/clean", methods=['GET', 'POST'])
def get_subscribers_clean_list_api():
    
    if request.method == 'GET':

        PUBLIC_KEY = request.args.get("public-key", None)
        PRIVATE_KEY = request.args.get("private-key", None)
        API_URL = request.args.get("api-url", None)
        payload = str(request.__dict__["environ"]["QUERY_STRING"])

        result = iterate_and_cleaning_up(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, payload)

        return jsonify(result)    
    
    else:
        
        PUBLIC_KEY = request.headers["public-key"]
        PRIVATE_KEY = request.headers["private-key"]
        API_URL = request.headers["api-url"]
        payload = str(request.form)
        
        values = list(request.form.values())
        keys = list(request.form.keys())
        
        params_string = ""
        for x, y in zip(values, keys):
            new_pair = "=".join([y,x+"&"])
            params_string = params_string + new_pair
        params_string = params_string[:-1]
        print(params_string)
    
        result = iterate_and_cleaning_up(INTERNAL_SERVER, PUBLIC_KEY, PRIVATE_KEY, API_URL, payload=params_string.strip())

        return jsonify(result)
```
